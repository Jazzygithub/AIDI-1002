```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.title('Final Math Grade Distribution Plot')
sns.distplot(StudMat.G3)

plt.subplot(1,2,2)
plt.title('Final Math Grade Spread')
sns.boxplot(y=StudMat.G3)

plt.show()

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.title('Final Portugese Grade Distribution Plot')
sns.distplot(StudPor.G3)

plt.subplot(1,2,2)
plt.title('Final Portugese Grade Spread')
sns.boxplot(y=StudPor.G3)

plt.show()
```
